```python
solution ="""
Giải: 
B1)Điểm trung bình, n = 5
x_ = (1+2+4+4+5) / 5 = 6/5 = 3.2
y_ = (2 + 4 + 5 + 4 + 6) / 5 = 4.2
B2) Tính S_xx và S_xy
| x_i | y_i | (x_i - x̄)  | (y_i - ȳ)  |   (x_i - x̄)(y_i - ȳ) |  (x_i - x̄)² |
|-----|-----|------------|------------|----------------------|-------------|
|  1  |  2  |    -2.2    |    -2.2    |         4.84         |      4.84   |
|  2  |  4  |    -1.2    |    -0.2    |         0.24         |      1.44   |
|  3  |  5  |     0.8    |     0.8    |         0.64         |      0.64   |
|  4  |  4  |     0.8    |    -0.2    |        -0.16         |      0.64   |
|  5  |  6  |     1.8    |     1.8    |         3.24         |      3.24   |
| Tổng|     |            |            |      S_xy = 8.8      | S_xx = 10.8 |

B3)Tính hệ số
w_1 = S_xy / S_xx = 8.8 / 10.8 = 0.8148
w_0 = y_ - w_1 * x_ = 4.2 - 0.8148 * 3.2 = 1.5926

vậy đường hồi quy: y_hat = 1.5926 + 0.8148x
B4)Dự đoán sinh viên học 6 giờ (x = 6)
y_hat = 1.5926 + 0.8148 * 6 = 6.48

Kết luận: Mỗi giờ tự học thêm giúp tăng thêm 0.8148 điểm; sinh viên tự học 6h được dự đoán
6.48 điểm
Độ khớp R^2 = 0.81
"""
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Dữ liệu
x = np.array([1,2,4,4,5])
y = np.array([2,4,5,4,6])
X = x.reshape(-1,1)
model = LinearRegression().fit(X, y)

w1 = model.coef_[0]
w0 = model.intercept_

print(f"y_hat = {w0:.4f} + {w1:.4f} * x")
print("Dự đoán x=6: ", round(model.predict([[6]])[0], 4), "điểm")

yhat = model.predict(X)

print("R^2: ", round(r2_score(y,yhat),4))
print("MSE: ", round(np.sqrt(mean_squared_error(y,yhat)),4))
```
